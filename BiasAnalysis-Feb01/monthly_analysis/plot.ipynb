```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
methods = ["IDW", "Kriging", "GP", "Geo-LSTM"]
performance = {
    method: {"RMSE": [], "CVRMSE": [], "MAE": [], "R2": []} for method in methods
}
```

# By Metric

```python
plt.plot(performance["IDW"]["RMSE"], label="IDW", marker="o")
plt.plot(performance["Kriging"]["RMSE"], label="Kriging", marker="o")
plt.plot(performance["GP"]["RMSE"], label="GP", marker="o")
plt.plot(performance["Geo-LSTM"]["RMSE"], label="Geo-LSTM", marker="o")
plt.xticks(range(4), ["Oct", "Nov", "Dec", "Jan"])
plt.xlabel("Month")
plt.ylabel("RMSE")
plt.title("Lower RMSE, Better Model")
plt.legend()
plt.show()
```

```python
plt.plot(performance["IDW"]["CVRMSE"], label="IDW", marker="o")
plt.plot(performance["Kriging"]["CVRMSE"], label="Kriging", marker="o")
plt.plot(performance["GP"]["CVRMSE"], label="GP", marker="o")
plt.plot(performance["Geo-LSTM"]["CVRMSE"], label="Geo-LSTM", marker="o")
plt.xticks(range(4), ["Oct", "Nov", "Dec", "Jan"])
plt.xlabel("Month")
plt.ylabel("CVRMSE")
plt.title("Lower CVRMSE, Better Model")
plt.legend()
plt.show()
```

```python
plt.plot(performance["IDW"]["MAE"], label="IDW", marker="o")
plt.plot(performance["Kriging"]["MAE"], label="Kriging", marker="o")
plt.plot(performance["GP"]["MAE"], label="GP", marker="o")
plt.plot(performance["Geo-LSTM"]["MAE"], label="Geo-LSTM", marker="o")
plt.xticks(range(4), ["Oct", "Nov", "Dec", "Jan"])
plt.xlabel("Month")
plt.ylabel("MAE")
plt.title("Lower MAE, Better Model")
plt.legend()
plt.show()
```

```python
plt.plot(performance["IDW"]["R2"], label="IDW", marker="o")
plt.plot(performance["Kriging"]["R2"], label="Kriging", marker="o")
plt.plot(performance["GP"]["R2"], label="GP", marker="o")
plt.plot(performance["Geo-LSTM"]["R2"], label="Geo-LSTM", marker="o")
plt.xticks(range(4), ["Oct", "Nov", "Dec", "Jan"])
plt.xlabel("Month")
plt.ylabel("R2")
plt.title("Higher R2, Better Model")
plt.legend()
plt.show()
```
