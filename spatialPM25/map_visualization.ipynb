```python
import geopandas as gpd
import matplotlib.pyplot as plt
from fiona.crs import from_string
import numpy as np

%matplotlib inline
```

```python
map = gpd.read_file("./map_data/ca-places-boundaries/CA_Places_TIGER2016.shp")
crs = from_string("+proj=longlat +datum=WGS84 +no_defs") 
map = map.to_crs(crs)
```

```python
tulare_map = map[map['NAME'] == "Tulare"]
tulare_map.plot()
```

```python
santaclara_map = map[map['NAME'] == "Santa Clara"]
santaclara_map.plot()
```
